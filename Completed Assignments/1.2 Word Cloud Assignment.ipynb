## Imports

We need `pandas` to load the CSV, `re` for regex based cleaning, `ENGLISH_STOP_WORDS` for filtering common words, and `WordCloud`/`matplotlib` for the visualization.

```python
import re
from collections import Counter

import matplotlib.pyplot as plt
import pandas as pd
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from wordcloud import WordCloud
```

## Step 1: Write the Cleaning Function

This function takes a raw line of dialogue and returns a cleaned version: lowercase, no punctuation, no stop words.

```python
def clean_text(text):
    """Lowercase, strip punctuation, and remove stop words from one line of text."""
    text = str(text).lower()
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    words = [w for w in text.split() if w and w not in ENGLISH_STOP_WORDS]
    return " ".join(words)
```

## Step 2: Load the Dataset

`movie-lines.csv` contains one row per line of dialogue spoken in a movie.

```python
df = pd.read_csv("../data/movie-lines.csv")
print(f"Loaded {len(df):,} movie lines")
df.head(3)
```

## Step 3: Clean the Text

Apply `clean_text` to every line, then join everything into one big corpus string for the word cloud.

```python
df["cleaned_text"] = df["text"].apply(clean_text)
corpus = " ".join(df["cleaned_text"])
```

## Step 4: Check Word Frequencies

Before visualizing, let's look at the raw numbers behind the word cloud.

```python
word_counts = Counter(corpus.split())

print("Top 15 most common words:")
for word, count in word_counts.most_common(15):
    print(f"  {word}: {count:,}")
```

## Step 5: Generate the Word Cloud

`min_word_length=2` filters out leftover single-letter fragments so the visualization stays readable.

```python
wordcloud = WordCloud(
    width=1000,
    height=500,
    background_color="white",
    colormap="viridis",
    max_words=150,
    min_word_length=2,
).generate(corpus)
```

## Step 6: Display the Word Cloud

```python
plt.figure(figsize=(14, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud \u2014 Cleaned Movie Dialogue (movie-lines.csv)")
plt.tight_layout()
plt.show()
```
